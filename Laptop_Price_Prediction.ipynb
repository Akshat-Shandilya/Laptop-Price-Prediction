## FEATURE ENGINEERING

### Company

```python
v1 = df['Company'].value_counts()
v1
```

```python
l1 = v1[v1<20].index
l1
```

```python
df['Company'] = df['Company'].replace( l1, 'Others' )
df['Company'].value_counts()
```

### TypeName

```python
df['TypeName'].value_counts()
```

### Screen Resolution

```python
df['ScreenResolution'].value_counts()
```

```python
df['IPS'] = df['ScreenResolution'].apply( lambda x : 1 if 'IPS' in x else 0 )
df['Touchscreen'] = df['ScreenResolution'].apply( lambda x : 1 if 'Touchscreen' in x else 0 )

df.head()
```

```python
df['temp'] = df['ScreenResolution'].str.split()
df.head()
```

```python
df['temp1'] = df['temp'].apply( lambda x : x[-1] )
df.head()
```

```python
df['temp2'] = df['temp1'].str.split('x')
df.head()
```

```python
df['X_res'] = df['temp2'].apply( lambda x : x[0] )
df['Y_res'] = df['temp2'].apply( lambda x : x[1] )
df.head()
```

```python
df['X_res'] = df['X_res'].astype('int')
df['Y_res'] = df['Y_res'].astype('int')
df.head()
```

### Size

```python
df['Size'] = ( df['X_res'] ** 2 + df['Y_res'] ** 2 ) ** 0.5
df.head()
```

```python
df = df.drop( columns = ['temp','temp1','temp2','X_res','Y_res','ScreenResolution'] )
df.head()
```

### CPU

```python
df['Cpu'].unique()
```

```python
def cputype(x) :
    if 'Intel' in x :
        if 'Intel Core i3' in x :
            return 'Intel Core i3' 
        elif 'Intel Core i5' in x :
            return 'Intel Core i5' 
        elif 'Intel Core i7' in x :
            return 'Intel Core i7'
        else :
            return'Intel Others'
    elif 'AMD' in x :
        return 'AMD' 
    else :
        return 'Others'
        
df['CPU'] = df['Cpu'].apply(cputype)
df['CPU'].unique()
```

```python
df = df.drop( columns = 'Cpu' )
df.head()
```

### RAM and Weight

```python
df['Ram'] = df['Ram'].str.replace('GB','')
df['Weight'] = df['Weight'].str.replace('kg','')

df['Ram'] = df['Ram'].astype('int')
df['Weight'] = df['Weight'].astype('float')

df.head()
```

### Memory

```python
df['Memory'].value_counts()
```

```python
df['Memory'] = df['Memory'].str.replace( 'TB', '000' )
df['Memory'] = df['Memory'].str.replace( 'GB', '' )
df['Memory'] = df['Memory'].str.replace( '.0', '' )
df['temp3'] = df['Memory'].str.split('+')
df.head()
```

```python
def SSDmemorytype (x) :
    for i in x :
        if 'SSD' in i :
            j = i.replace('SSD','')
            k = j.replace(' ','')
            k = int(k)
            return k
df['SSD'] = df['temp3'].apply(SSDmemorytype)

def HDDmemorytype (x) :
    for i in x :
        if 'HDD' in i :
            j = i.replace('HDD','')
            k = j.replace(' ','')
            k = int(k)
            return k
df['HDD'] = df['temp3'].apply(HDDmemorytype)

def Hybridmemorytype (x) :
    for i in x :
        if 'Hybrid' in i :
            j = i.replace('Hybrid','')
            k = j.replace(' ','')
            k = int(k)
            return k
df['Hybrid'] = df['temp3'].apply(Hybridmemorytype)

def FlashStoragememorytype (x) :
    for i in x :
        if 'Flash Storage' in i :
            j = i.replace('Flash Storage','')
            k = j.replace(' ','')
            k = int(k)
            return k
df['Flash Storage'] = df['temp3'].apply(FlashStoragememorytype)

df['SSD'] = df['SSD'].fillna(0)
df['HDD'] = df['HDD'].fillna(0)
df['Hybrid'] = df['Hybrid'].fillna(0)
df['Flash Storage'] = df['Flash Storage'].fillna(0)

df = df.drop( columns = ['Memory','temp3'] )

df.head()
```

### GPU

```python
df['Gpu'].unique()
```

```python
def gputype(x) :
    if 'Intel' in x :
        if 'Intel Iris Plus Graphics' in x :
            return 'Intel Iris Plus Graphics' 
        elif 'Intel HD Graphics' in x :
            return 'Intel HD Graphics' 
        else :
            return'Intel Others'
    elif 'Nvidia' in x :
        if 'Nvidia Quadro' in x :
            return 'Nvidia Quadro' 
        elif 'Nvidia GeForce' in x :
            return 'Nvidia GeForce' 
        else :
            return'Nvidia Others'
    elif 'AMD Radeon' in x :
        return 'AMD Radeon'
    else :
        return 'Others'

df['GPU'] = df['Gpu'].apply(gputype)
df['GPU'].unique()
```

```python
df = df.drop( columns=['Gpu'] )
df.head()
```

### Operating 

```python
df['OpSys'].value_counts()
```

## Encoding

```python
df.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer( transformers = [
        ( 't1', OneHotEncoder( drop='first' ), ['Company','TypeName','OpSys','CPU','GPU'] )
    ] , remainder='passthrough' )
```

```python
df_new = pd.DataFrame( ct.fit_transform(df) )
df_new.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_new.drop(columns=[36]), df_new.iloc[:,36], test_size=0.2 )
x_train.head()
```

```python
y_train.head()
```

## TRAINING MODELS

### Linear Regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, r2_score, mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred_LR = lr.predict( x_test )
```

```python
from sklearn.metrics import mean_absolute_error
print (mean_absolute_error( y_pred_LR, y_test ))
print (r2_score( y_pred_LR, y_test ))
```

### K Neighbors Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(x_train, y_train)
y_pred_KNN = lr.predict( x_test )
```

```python
print (mean_absolute_error( y_pred_KNN, y_test ))
scores_knn= cross_val_score(knn, x_test, y_test, cv=10, scoring=make_scorer(r2_score))
print( scores_knn.mean() )
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor( max_depth=10 )
dtr.fit( x_train, y_train )
y_pred_DTR = dtr.predict( x_test )
```

```python
print (mean_absolute_error( y_pred_DTR, y_test ))
scores_knn = cross_val_score(knn, x_test, y_test, cv=10, scoring=make_scorer(r2_score))
print( scores_knn.mean() )
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)
rfr.fit( x_train, y_train )
y_pred_RFR = rfr.predict( x_test )
```

```python
print (mean_absolute_error( y_pred_RFR, y_test ))
scores_rfr= cross_val_score(rfr, x_test, y_test, cv=10, scoring=make_scorer(r2_score))
print( scores_rfr.mean() )
```

### Voting Regressor

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, VotingRegressor
from xgboost import XGBRegressor

rfr = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15, bootstrap=True)
gbr = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
xgbr = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)
etr = ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=10, bootstrap=True)

vr = VotingRegressor([('rfr', rfr), ('gbr', gbr), ('xgbr', xgbr), ('etr', etr)],
                     weights=[5, 1, 1, 1])

vr.fit( x_train, y_train )
y_pred_VR = vr.predict( x_test )
```

```python
print (mean_absolute_error( y_pred_VR, y_test ))
scores_vr= cross_val_score(vr, x_test, y_test, cv=10, scoring=make_scorer(r2_score))
print( scores_vr.mean() )
```

### Neural networks

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, r2_score

model = Sequential([
    Dense(128, input_dim=43, activation='relu'),  # Input layer with 128 nodes
    Dropout(0.2),  # Dropout layer to reduce overfitting
    Dense(64, activation='relu'),  # Hidden layer with 64 nodes
    Dense(32, activation='relu'),  # Hidden layer with 32 nodes
    Dense(1)  # Output layer for regression
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
y_pred_cnn = model.predict(x_test)
```

```python
print( r2_score(y_test, y_pred_cnn) )
```
